```python
def rotImage(image,angle,retain):
    angle=math.radians(angle)                               
    cosine=math.cos(angle)
    sine=math.sin(angle)

    height=image.shape[0]                                   
    width=image.shape[1]                                    
    if(retain==0):
        new_height  = round(abs(height*cosine)+abs(image.shape[1]*sine))+1
        new_width  = round(abs(width*cosine)+abs(image.shape[0]*sine))+1
    else:
        new_height = height
        new_width = width
        

    output=np.zeros((new_height,new_width,image.shape[2]))

    # Centre of the image about which we have to rotate the image
    original_centre_height   = round(((height+1)/2)-1)    
    original_centre_width    = round(((width+1)/2)-1)    
    old_centre_height = original_centre_height
    old_centre_width = original_centre_width
    
    # Centre of the new image that will be obtained
    new_centre_height= round(((new_height+1)/2)-1)        
    new_centre_width= round(((new_width+1)/2)-1)     
    a= np.arange(width)
    b= np.arange(height)
    f1 = interpolate.interp2d(a, b, image[:,:,0])
    f2 = interpolate.interp2d(a, b, image[:,:,1])
    f3 = interpolate.interp2d(a, b, image[:,:,2])

    for i in range(new_height):
        for j in range(new_width):
            #co-ordinates of pixel with respect to the centre of original image
            
            y= new_height-1-i-new_centre_height                   
            x= new_width-1-j-new_centre_width                      

            #co-ordinate of pixel with respect to the old rotated image
            
            old_y=(-x*sine+y*cosine)
            old_x=(x*cosine+y*sine)
            
            #need to change old_x and old_y with respect to the old centre
            
            old_y=old_centre_height-old_y
            old_x=old_centre_width-old_x
            
            # Checking if the index is valid or not
            
            if 0 <= old_x < width and 0 <= old_y < height and old_x>=0 and old_y>=0:
                output[i,j,0]=f1(old_x,old_y)
                output[i,j,1]=f2(old_x,old_y)
                output[i,j,2]=f3(old_x,old_y)
    return output.astype(np.uint8)
```

```python
image=cv.imread('../images/lg.jpg',1)
image=cv.cvtColor(image,cv.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(image)
plt.title("Original Image")
```

### 90 degrees with and without cropping

```python
res1 = rotImage(image,90,1)
res2 = rotImage(image,90,0)
f, axarr = plt.subplots(1,2,figsize=(15,10))
axarr[0].imshow(res1)
axarr[1].imshow(res2)
axarr[0].set_title("With Cropping")
axarr[1].set_title("Without Cropping")
```
