```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
def gammaTransform(im,gamm):
    return (255*((im/255)**gamm)).astype(np.uint8)
```

```python
st_img=cv.imread("../images/stealth.png")
st_img=cv.cvtColor(st_img,cv.COLOR_BGR2GRAY)
st_gam = gammaTransform(st_img,0.4)

f, axarr = plt.subplots(1,2,figsize=(15,10))
axarr[0].imshow(st_gam,cmap="gray")
axarr[1].imshow(st_img,cmap="gray")
axarr[0].set_title("Gamma Corrected with gamma 0.4")
axarr[1].set_title("Original stealth image")
```

```python
unst_img= cv.imread('../images/unstealth.png',0)
unst_gam=gammaTransform(unst_img,2)
f, axarr = plt.subplots(1,2,figsize=(15,10))
axarr[0].imshow(unst_gam,cmap="gray")
axarr[1].imshow(unst_img,cmap="gray")
axarr[0].set_title("Gamma Corrected with gamma 2")
axarr[1].set_title("Original unstealth image")
```

```python
aston_img= cv.imread('../images/aston_invisible.jpg')
aston_img =cv.cvtColor(aston_img,cv.COLOR_BGR2GRAY)
aston_gam = gammaTransform(aston_img,0.35)
f, axarr = plt.subplots(1,2,figsize=(15,10))
axarr[0].imshow(aston_gam,cmap="gray")
axarr[1].imshow(aston_img,cmap="gray")
axarr[0].set_title("Gamma Corrected with gamma 0.35")
axarr[1].set_title("Original Invisible Aston image")
```
