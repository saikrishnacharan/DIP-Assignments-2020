```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def linContrastStretching(img,a,b):
    mini=np.amin(img)
    maxi=np.amax(img)
    return np.round(a + (b-a)* ((img-mini)/(maxi-mini)))
```

```python
img=cv.imread('../images/quantum_bad.jpg',1)
img_gs=cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
mini=np.amin(img_gs)
maxi=np.amax(img_gs)

print("Original Image minimum Value: ",mini)
print("Original Image maximum Value: ",maxi)
```

```python
kk=linContrastStretching(img_gs,0,255)
print("Contrasted Image minimum Value: ",np.amin(kk))
print("Contrasted Image maximum Value: ",np.amax(kk))
```

```python
f, axarr = plt.subplots(2,2,figsize=(20,10))
axarr[0,1].imshow(kk.astype('uint8'))
axarr[0,0].imshow(img_gs)
axarr[0,1].set_title("Contrasted Image")
axarr[0,0].set_title("Original Image")
axarr[1,0].hist(img_gs.ravel(), bins=256, range=(0.0, 255.0))
axarr[1,1].hist(kk.ravel(), bins=256, range=(0.0, 255.0))
axarr[1,0].set_title("Original Image Histogram")
axarr[1,1].set_title("Contrasted Image Histogram")
```

- The contrast improvement is done by bringing all the pixel values to the range [a,b]. 
- We normalize all the pixel values to [0,1] such that min pixel value is mapped to 0 and max pixel value is mapped to 1.   The equation corresponding is  $\frac{(im - np.amin(im))}{(np.amax(im)- np.amin(im)}$
- $np.amin()$ is the min intensity value while $np.amax()$ is the maximum intensity value
- Now, this is multiplied by b-a to bring the range to [0, b-a]
- And, a is added to this to get the range to [a , b]
- In this corresponding image what we did is we increased the range of image from [0 82] to [0 255] which makes look visually better than original image
